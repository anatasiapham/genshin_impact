# **Data Wrangling/Data Cleaning**

---



```python
# import data
import pandas as pd
df = pd.read_csv('genshin.csv')
```

```python
df.head()
```

```python
df.describe(include='all')
```

```python
df.columns
```

```python
df = df.drop(columns=['hp_80_90', 'atk_80_90', 'def_80_90',
       'hp_80_80', 'atk_80_80', 'def_80_80', 'hp_70_80', 'atk_70_80',
       'def_70_80', 'hp_70_70', 'atk_70_70', 'def_70_70', 'hp_60_70',
       'atk_60_70', 'def_60_70', 'hp_60_60', 'atk_60_60', 'def_60_60',
       'hp_50_60', 'atk_50_60', 'def_50_60', 'hp_50_50', 'atk_50_50',
       'def_50_50', 'hp_40_50', 'atk_40_50', 'def_40_50', 'hp_40_40',
       'atk_40_40', 'def_40_40', 'hp_20_40', 'atk_20_40', 'def_20_40',
       'hp_20_20', 'atk_20_20', 'def_20_20', 'hp_1_20', 'atk_1_20', 'def_1_20',
       'spsecial_0', 'special_1', 'special_2', 'special_3', 'special_4',
       'special_5', 'special_6'])
```

```python
# check missing values
df.isnull().sum()/len(df)
```

```python
# check for NaNs in the data set
df[df.isna().any(axis=1)]
# I notice most of the NAs are from characters that are not yet added into the game and regions for the rest
```

```python
df = df.drop(labels=[1,5,13,14,34], axis=0) # dropped characters that were not yet playable (not yet added into the game)
```

```python
# the regions of characters Aloy, Aether, and Lumine are in fact unknown in the lore, so instead of NA I will change it to be 'unknown'
df.loc[[2,50,51],['region']] = 'Unknown'

# double check that it was successful
df['region'].value_counts()
```

```python
# double check for no more NaNs
df[df.isna().any(axis=1)]
```

```python
# convert to release_date to datetime 
df["release_date"] = pd.to_datetime(df['release_date']).dt.date
df["year"] = pd.to_datetime(df['release_date']).dt.year
df["month"] = pd.to_datetime(df['release_date']).dt.month
df["month_name"] = pd.to_datetime(df['release_date']).dt.month_name()


df.head()
```

# **Data Visualization**

---



#**Region**

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
regions=sns.countplot(data=df, x="region",order=df['region'].value_counts().index, hue="region", dodge=False)

# This will add title to plot
regions.set_title("Character Count by Region")
 
# This will add label to X-axis
regions.set_xlabel("Genshin Regions")

# This will add label to Y-axis
regions.set_ylabel("Count")
```

### **According to the plot by Genshin regions, the region with the most characters was Mondstadt. Interestingly, the one character from Snezhnaya was Tartaglia.**

#**Month**

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

sns.set(rc={'figure.figsize':(11.7,8.27)})
monthly=sns.countplot(data=df, x="month_name", order=months)

# This will add title to plot
monthly.set_title("Total Character Count by Month")
 
# This will add label to X-axis
monthly.set_xlabel("Months")
monthly.set_xticklabels(monthly.get_xticklabels(), rotation=15)

# This will add label to Y-axis
monthly.set_ylabel("Count")
```

### **After organizing the release months of each character, it was shocking to find that most Genshin characters were released in September. I wonder if there is a reason for this?**

#**Year**

```python
df_filtered = df.groupby(['year','month'])['character_name'].count().reset_index(name='counts')
df_filtered["month_year"] = df_filtered["year"].astype(str) + "-" + df_filtered["month"].astype(str)
```

```python
all_years = sns.relplot(
    data=df_filtered,
    x="month", y="counts",
    hue="year",
    kind="line",
    height=5, aspect=1, facet_kws=dict(sharex=False),
)
all_years.fig.suptitle('Total Characters for All Years', fontsize=16)
all_years.fig.subplots_adjust(top=0.9)
```

```python
by_year = sns.FacetGrid(df_filtered, col="year", margin_titles=True, col_wrap=3, height=4, aspect=1, hue="year")
by_year.map_dataframe(
    sns.lineplot, x="month", y="counts"
)
by_year.fig.suptitle('Total Characters by Year', fontsize=16)
by_year.fig.subplots_adjust(top=0.85)
```

### **After investigating further through more visualizations, I learned that September for the previous chart had a large spike because that was the release date of Genshin and hence why numerous characters were released that month during 2020.**

### Even after accounting for the spike from the initial launch of the game, September still saw the most characters released over the years.

# **4-Star vs 5-Star**
